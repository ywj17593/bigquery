```python
# Sampling data from job_log
from google.cloud import bigquery

client = bigquery.Client()
```

```python
sql = """
SELECT * 
FROM `macro-centaur-330222.service_migration_dataset.cluster_job_log` 
"""
df = client.query(sql).to_dataframe()
df.head()
```

```python
import pandas as pd
data_summary=pd.DataFrame(columns=['vc','job_id', 'cpu_id','gpu_id','cpu_max','cpu_min','cpu_ave','gpu_max','gpu_min','gpu_ave','mem_max','mem_min','mem_ave','status'])
data_summary
```

```python
import time
sample=df.sample(1)
sample
```

```python
start_time=sample.iloc[0].utc_start_time
end_time=sample.iloc[0].utc_end_time
cpu_id=sample.iloc[0].cpu_id
gpu_id=sample.iloc[0].gpu_id
start_time=start_time.strftime("%Y-%m-%d %H:%M:%S")
end_time=end_time.strftime("%Y-%m-%d %H:%M:%S")
start_time,end_time,cpu_id,gpu_id
```

```python
# Sharable CPU

sql_share = """ SELECT * FROM `macro-centaur-330222.service_migration_dataset.cluster_job_log` WHERE cpu_id=\"""" + cpu_id + """\" AND ((utc_start_time >=TIMESTAMP (\"""" + start_time + """\", \"UTC\") AND utc_start_time <=TIMESTAMP (\"""" + end_time + """\", \"UTC\")) OR (utc_end_time <=TIMESTAMP (\"""" + end_time + """\", \"UTC\")AND utc_end_time >=TIMESTAMP (\"""" + start_time + """\", \"UTC\")))"""
sharable = client.query(sql_share).to_dataframe()
sharable
```

```python
# cpu_usage
sql_cpu = """ SELECT * FROM `macro-centaur-330222.service_migration_dataset.cluster_cpu_util` WHERE machine_id=\"""" + cpu_id + """\" AND time_utc >TIMESTAMP (\"""" + start_time + """\", \"UTC\") AND time_utc <TIMESTAMP (\"""" + end_time + """\", \"UTC\")"""
cpu_usage = client.query(sql_cpu).to_dataframe()
cpu_usage.head()
```

```python
# gpu_usage
sql_gpu = """ SELECT machineId,"""+gpu_id+"""_util,time_utc FROM `macro-centaur-330222.service_migration_dataset.cluster_gpu_util` WHERE machineId=\"""" + cpu_id + """\" AND time_utc >TIMESTAMP (\"""" + start_time + """\", \"UTC\") AND time_utc <TIMESTAMP (\"""" + end_time + """\", \"UTC\")"""
gpu_usage = client.query(sql_gpu).to_dataframe()
gpu_usage.head()
```

```python
# mem_usage
sql_mem = """ SELECT * FROM `macro-centaur-330222.service_migration_dataset.cluster_mem_util` WHERE machine_id=\"""" + cpu_id + """\" AND time_utc >TIMESTAMP (\"""" + start_time + """\", \"UTC\") AND time_utc <TIMESTAMP (\"""" + end_time + """\", \"UTC\")"""
mem_usage = client.query(sql_mem).to_dataframe()
mem_usage.insert(3,'mem_util',(1-(mem_usage.mem_free/mem_usage.mem_total))*100)
mem_usage.head()
```

```python
#Analysis of data
new_record=pd.DataFrame([[sample.iloc[0].vc,sample.iloc[0].jobid, sample.iloc[0].cpu_id,sample.iloc[0].gpu_id,cpu_usage.cpu_util.max(),cpu_usage.cpu_util.min(),cpu_usage.cpu_util.mean(),gpu_usage.iloc[:,1].max(),gpu_usage.iloc[:,1].min(),gpu_usage.iloc[:,1].mean(),mem_usage.mem_util.max(),mem_usage.mem_util.min(),mem_usage.mem_util.mean(),sample.iloc[0].status]],columns=data_summary.columns)
data_summary=data_summary.append(new_record,ignore_index=True)
data_summary
```

```python
# Mapping cluster with machine

sql_map = """ SELECT DISTINCT(cpu_id), vc, jobid,utc_submitted_timeacro-centaur-330222.service_migration_dataset.cluster_job_log` ORDER BY cpu_id"""
mapping = client.query(sql_map).to_dataframe()
mapping
```

```python
sql_map = """ SELECT COUNT(DISTINCT(cpu_id)), vc FROM `macro-centaur-330222.service_migration_dataset.cluster_job_log` GROUP BY vc"""
mapping = client.query(sql_map).to_dataframe()
mapping
```

```python
sql_map = """ SELECT COUNT(DISTINCT(jobid)), status FROM `macro-centaur-330222.service_migration_dataset.cluster_job_log` GROUP BY status"""
mapping = client.query(sql_map).to_dataframe()
mapping
```
