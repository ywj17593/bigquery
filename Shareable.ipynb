```python
%load_ext google.cloud.bigquery
# Sampling data from job_log
from google.cloud import bigquery

client = bigquery.Client()
sql = """
SELECT DISTINCT(jobid),utc_start_time,utc_end_time, cpu_id 
FROM `macro-centaur-330222.service_migration_dataset.cluster_job_log` 
"""
df = client.query(sql).to_dataframe()
df
```

```python
import pandas as pd
import time
count=0
allcount=0
for i in range(10000):
    start_time=df.iloc[i].utc_start_time
    end_time=df.iloc[i].utc_end_time
    cpu_id=df.iloc[i].cpu_id
    for j in df.index:
        if df["cpu_id"][j]==cpu_id and ((df["utc_start_time"][j]>start_time and df["utc_start_time"][j]<end_time)or(df["utc_end_time"][j]>start_time and df["utc_end_time"][j]<end_time)):
            count+=1
            break
    allcount+=1                        
count/allcount

```
